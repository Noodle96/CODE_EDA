```python
def multiply_mod(M, N, m):
    a = ((((M[0][0]%m)*(N[0][0]%m))%m)%m + (((M[0][1]%m)*(N[1][0]%m)%m)%m)%m)
    b = ((((M[0][0]%m)*(N[0][1]%m))%m)%m + (((M[0][1]%m)*(N[1][1]%m)%m)%m)%m)
    c = ((((M[1][0]%m)*(N[0][0]%m))%m)%m + (((M[1][1]%m)*(N[1][0]%m)%m)%m)%m)
    d = ((((M[1][0]%m)*(N[0][1]%m))%m)%m + (((M[1][1]%m)*(N[1][1]%m)%m)%m)%m)

    N[0][0] = a
    N[0][1] = b
    N[1][0] = c
    N[1][1] = d

def power_mod(M,n,m):
    if n == 1 or n == 0:
        return
    if n % 2 == 0:
        power_mod(M, n//2, m)
        multiply_mod(M, M, m)
    else:
        A = [[0%m,1%m],[1%m,1%m]]
        power_mod(M, n//2, m)
        multiply_mod(M, M, m)
        multiply_mod(A, M, m)

def fib_mod(n,m):
    if n <= 1:
        return n%m
    M = [[0%m,1%m],[1%m,1%m]]
    power_mod(M,n-1, m)
    return M[1][1]
```
